```python
# Importando a biblioteca seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando a base de dados 'titanic'
titanic = sns.load_dataset("titanic")

titanic.head()
```

### Sobre os dados


No exercício de hoje, vamos utilizar as variáveis:

- `age`: idade do passageiro
- `fare`: tarifa do passageiro



### Exercício de hoje - Parte 1

Utilize a função `sns.lineplot()` para criar um gráfico de linha que exiba a tarifa média a partir da idade do passageiro, e remova também a barra de erro com o argumento `errorbar=None`.

Use a idade no eixo X e a tarifa no eixo Y.


Após escrever a linha de código de sua resposta, execute a célula para mostrar o gráfico a partir da base `titanic`.

**Importante: antes de executar o código da sua resposta, não deixe de executar a célula de código que está no início deste caderno, que importa as bibliotecas e carrega os dados.**

```python
## Escreva sua resposta abaixo e execute a célula para gerar o gráfico
sns.lineplot(
    x="age",
    y="fare",
    errorbar=None,
    estimator="mean", # essa linha é opcional
    data=titanic,
)
```

```python
# Continue o código abaixo
fig, ax = plt.subplots(figsize=(12, 6))

sns.lineplot(
    x="age",
    y="fare",
    errorbar=None,
    estimator="mean", # essa linha é opcional
    data=titanic,
)

ax.set_title("Relação entre idade e tarifa")
ax.set_xlabel("Idade")
ax.set_ylabel("Tarifa (USD)")
```
