## 1.2. Vamos contar categorias!

Agora que já carregamos as bibliotecas e temos uma base de dados para trabalhar, já podemos criar o nosso primeiro gráfico!

Vamos começar com um gráfico simples, que nos permita contar quantas transações foram feitas a partir do número de linhas em que uma variável categórica específica aparece. Para isso vamos usar a função `countplot()` da biblioteca `seaborn`.

### 1.2.1. Argumentos básicos da função `countplot`

A função `countplot` é utilizada quando queremos construir um gráfico de barras simples por meio da contagem de categorias em uma de nossas variáveis.

Sendo assim, vamos começar analisando o número de transações feitas em cada forma de pagamento (`"Crédito"` ou `"Débito"`). Basta indicar o eixo em que queremos representar os nomes das categorias, com o nome da respectiva coluna entre aspas. Vamos colocar as categorias de forma de pagamento no eixo `x`, passando para a função `x="forma_pagamento"`. Por fim, precisamos também especificar a base de dados que estamos utilizando, ou seja, de onde está saindo a coluna `"forma_pagamento"`. Indicamos que queremos usar a base de dados `cartoes` passando para a função `data=cartoes`. Com isso chegamos à linha de código a seguir:

````python
# Criando nosso primeiro gráfico!
sns.countplot(x="forma_pagamento", data=cartoes)
````

Pronto! Com isso já geramos o nosso primeiro gráfico. Percebemos que os pagamentos do cartão presidencial, no período analisado, foram feitos com muito mais frequência em crédito do que em débito.

### 1.2.2. Invertendo os eixos

Agora, como poderíamos mudar um pouco esse gráfico para indicar as categorias de forma de pagamento no eixo y e a contagem do número de transações no eixo x (invertendo o gráfico acima)? Ora, bastaria vincular a nossa variável categórica ao eixo y em vez do x. Podemos substituir o argumento `x="forma_pagamento"` passando `y="forma_pagamento"` para a função `countplot()`, como se vê a seguir:

````python
sns.countplot(y="forma_pagamento", data=cartoes)
````

Simples, não? Já conseguimos criar nosso primeiro gráfico de contagem de frequência e ainda temos uma forma simples de escolher em qual eixo queremos representar nossa contagem. Mas talvez você esteja por que seria útil escolher o eixo como fizemos agora.

Para reponder a essa pergunta, vamos criar um novo gráfico, dessa vez utilizando outra variável categórica presente nos nossos dados: o `mandato` presidencial de cada transação. No código a seguir, começamos novamente indicando os mandatos no eixo x e informando que estamos usando os dadso da variável `cartoes`:

````python
sns.countplot(x="mandato", data=cartoes)
````

O que você acha desse gráfico? Conseguimos ler bem o nome dos mandatos no eixo x? Não, definitivamente não está legível. Os nomes das categorias estão sobrepostos e torna-se muito difícil distinguir as categorias representadas no gráfico.

Isso acontece porque a nossa linguagem se escreve da esquerda para a direita, ou seja, no mesmo sentido do eixo x de um gráfico. Por conta disso, sempre que temos um grande número de categorias, ou nomes extensos das categorias, é comum que surjam problemas de sobreposição dos nomes das categorias. Uma forma bem simples de lidar com isso (existem outras que veremos mais à frente) é mudar o eixo em que essas categorias estão representadas. Veja como fica o nosso gráfico dos mandatos quando invertemos os eixos:

````python
sns.countplot(y="mandato", data=cartoes)
````

Agora podemos ver com facilidade os nomes dos mandatos e a contagem dos pagamentos. Sabendo disso, de acordo com o nosso gráfico: 

<center><b>Qual presidente gastou mais durante o seu mandato?</center></b>

Você pensou imediatamente que o presidente que gastou mais foi o Lula em seu segundo mandato? **Tome cuidado!** O nosso gráfico não permite tirar essa conclusão. Não sabemos o valor das transações, apenas o número total de transações. O gráfico está dizendo, apenas, que o maior número de transações foi realizado durante o segundo mandato do presidente Lula. No entanto, se o valor médio dessas transações for inferior ao de outros mandatos, é possível que o gasto total nesse mandato não tenha sido o mais alto (veremos no próximo capítulo que, de fato, não é!).

Esta pergunta está aqui como aviso de algo que será fundamental para você sempre que estiver analisando dados: tenha clareza do que está sendo medido e como. Se atente para a unidade de análise utilizada e sempre busque saber exatamente como cada dimensão do gráfico está sendo representada.

### 1.2.3. A função `countplot` nunca aprendeu a somar :(

Nós temos uma variável em nossa base de dados que indica o valor de cada transação. E aí, será que conseguimos utilizá-la com a função `countplot` para calcular o gasto total de cada mandato presidencial?

Infelizmente, a resposta nesse caso é não. A função `countplot` não sabe agregar uma variável numérica para cruzá-la com uma variável categórica. Se tentamos passar uma variável numérica para o eixo x do nosso gráfico, junto com nossa variável categórica no eixo y, obteremos uma mensagem de erro dizendo que só podemos usar um único eixo com a função `countplot`. Veja a última linha da mensagem de erro gerada pelo código a seguir:


````python
sns.countplot(y="mandato", x="valor", data=cartoes)
````

Mas, não se preocupe! O `seaborn` possui uma função que sabe tirar médias, somar ou fazer ainda inúmeras outras operações de agregação de uma variável numérica para relacioná-la a uma variável categórica. Trata-se da função `barplot`, que conheceremos no próximo capítulo desta apostila. Aguarde!

### 1.2.4. Usando cores para cruzar categorias

Vamos encerrar este primeiro capítulo aprendendo a cruzar duas variáveis categóricas, que serão representadas como barras de cores diferentes em nosso gráfico. Para isso, basta utilizar o argumento `hue`, que associada cada categoria a uma cor e cria uma legenda em nosso gráfico para indicar as cores. Precisamos definir de qual coluna dos nossos dados vamos extrair as categorias que serão representadas como cores distintas no gráfico.

Observe o código a seguir. Nele estamos criando um gráfico de contagem de transações por mandato, colocando as categorias de mandato no eixo y (`y="mandato"`). Além disso, estamos contando separadamente, em cada mandato, o número de transações realizadas em cada forma de pagamento, crédito ou débito (`hue="forma_pagamento"`).

````python
sns.countplot(
    y="mandato",
    hue="forma_pagamento",
    data=cartoes
)
````

O gráfico nos permite observar que houve uma clara tendência de redução do número de transações realizadas em débito ao longo dos mandatos presidenciais. Essa tendência, inclusive, parece se confirmar mesmo quando o número total de transações cresce de um mandato para o outro. O gráfico que criamos já é capaz de evidenciar, portanto, uma relação complexa entre as informações disponíveis acerca dos pagamentos, algo que dificilmente conseguiríamos identificar facilmente apenas trabalhando diretamente os dados tabulares, sem produzir visualizações.

Para mostrar um pouco das customizações que aprenderemos a fazer ao longo do curso, deixamos a seguir um exemplo de um código mais "bem acabado" para produzir o último gráfico que havíamos mostrado, que conta o número de transações por forma de pagamento, por mandato.

Não se preocupe em entender tudo o que acontece nesse código. Veremos com mais calma, ao longo do curso, como funcionam essas customizações.

````python
fig, ax = plt.subplots()

# Contagem de transações por forma de pagamento e por mandato
sns.countplot(
    y="mandato",
    hue="forma_pagamento",
    palette={"Crédito": "#1e4ce3", # indicação da palheta de cores
             "Débito": "#1fdb51"},
    data=cartoes,
    ax=ax
)

# Adicionar título ao gráfico
ax.set_title("Volume de transações dos cartões presidenciais\n(por mandato e forma de pagamento)")

# Especificar o nome do eixo x
ax.set_xlabel("Número de Transações")

# Remover o nome do eixo y
ax.set_ylabel("")

# Especificar o título da legenda
ax.legend(title="Forma de Pagamento")

plt.show()
````
